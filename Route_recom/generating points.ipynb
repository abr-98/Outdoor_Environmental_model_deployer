```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
leftup=23.565774, 87.269568
rightdown=23.534924, 87.321653
```

```python
axis_y=np.linspace(rightdown[0],leftup[0],16)
axis_x=np.linspace(leftup[1],rightdown[1],16)
```

```python
X,Y=np.meshgrid(axis_x,axis_y)
```

```python
plt.scatter(X.flatten(),Y.flatten())
plt.xlim(87.265,87.325)
plt.ylim(23.53,23.57)
```

```python
list(zip(X.flatten(),Y.flatten()))
```
